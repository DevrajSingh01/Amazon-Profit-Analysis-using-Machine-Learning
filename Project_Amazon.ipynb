```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Amazon_df = pd.read_csv("Amazon 2_Raw.csv")
Amazon_df
```

```python
Amazon_df["Category"].value_counts()
```

```python
Amazon_df["Product Name"].value_counts()
```

```python
Amazon_df.groupby("Category")["Profit"].sum().plot(kind="bar", color=["blue", "purple", "orange", "gray", "red", "green", "yellow", "pink", "cyan", "magenta", "turquoise", "indigo", "violet", "maroon", "teal", "gold", "silver"])
plt.title("Profit per category")
plt.xlabel("Category")
plt.ylabel("Profit")
plt.show()
```

```python
remove_Bookcases = Amazon_df[Amazon_df["Category"] == "Bookcases"].index
remove_Bookcases
```

```python
Amazon_subset = Amazon_df.drop(remove_Bookcases)
Amazon_subset
```

```python
Amazon_info = Amazon_subset[
    ["Order ID", "Order Date", "Ship Date", "Category", "Product Name", "Sales", "Quantity", "Profit"]
]
Amazon_info
```

```python
colors = []
for index, row in Amazon_info.iterrows():
    Category = row["Category"]
    if Category == "Accessories":
        colors.append("blue")
    elif Category == "Phones":
        colors.append("purple")
    elif Category == "Appliances":
        colors.append("orange")
    else:
        colors.append("gray")
```

```python
fig = plt.figure(figsize=(12, 8))
plt.scatter(
    Amazon_info["Sales"], Amazon_info["Profit"], c=colors, marker="o"
)
plt.title("Sales vs Profit in electronic product")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.show()
```
